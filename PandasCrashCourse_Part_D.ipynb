# 2. Combining Boolean filtering

```python
over_60= df['age']>60
cabin_exists= df['cabin'].notnull()
```

```python
over_60_cabin_exists = over_60+ cabin_exists
```

```python
#the above can be written as 

over_60_cabin_exists = over_60 | cabin_exists
```

```python
over_60_cabin_exists
```
